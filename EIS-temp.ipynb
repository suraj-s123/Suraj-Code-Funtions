Fretchet Distance

```python
from scipy.spatial.distance import directed_hausdorff 
from scipy.spatial import distance  

def frechet_distance(P, Q): 
    ca = np.full((len(P), len(Q)), -1.0)
    def c(i, j):
        if ca[i, j] > -1:
            return ca[i, j]
        elif i == 0 and j == 0:
            ca[i, j] = distance.euclidean(P[0], Q[0]) 
        elif i>0 and j==0:
            ca[i, j] = max(c(i-1, 0), distance.euclidean(P[i], Q[0])) 
        elif i==0 and j >0:
            ca[i, j] = max(c(0, j-1), distance.euclidean(P[0], Q[j])) 
        elif i>0 and j>0:
            ca[i, j] = max(
                min(c(i-1, j), c(i-1, j-1), c(i, j-1)), distance.euclidean(P[i], Q[j])
            )
        return ca[i, j]
    return c(len(P) - 1, len(Q) - 1)

frechet_scores = [] 

for i, row in enumerate(asoh.itertuples(index = False)): 
    x_values = row[:13] 
    y_values = row[13:26] 

    x_values_normalized = [(x - x_min) / (x_max - x_min) for x in x_values] 
    y_values_normalized = [(y - y_min) / (y_max - y_min) for y in y_values] 
    y_values_normalized = [-1*i for i in y_values_normalized] 

    curve = np.column_stack((x_values_normalized, y_values_normalized)) 
    
    if i == 0:
        reference_curve = curve 
    frechet_dist = frechet_distance(curve, reference_curve)
    frechet_scores.append(frechet_dist)
    plt.plot(x_values_normalized, y_values_normalized, marker='x')

plt.title("Normalized Curves")
plt.xlabel("X Normalized")
plt.ylabel("Y Normalized")
plt.show()
print(f"Mean Fréchet Distance: {np.mean(frechet_scores)}")
```

```python
from fastdtw import fastdtw
from scipy.spatial import distance
from scipy.spatial.distance import directed_hausdorff

# Function to compute Fréchet distance
def frechet_distance(P, Q):
    ca = np.full((len(P), len(Q)), -1.0)
    
    def c(i, j):
        if ca[i, j] > -1:
            return ca[i, j]
        elif i == 0 and j == 0:
            ca[i, j] = distance.euclidean(P[0], Q[0])
        elif i > 0 and j == 0:
            ca[i, j] = max(c(i-1, 0), distance.euclidean(P[i], Q[0]))
        elif i == 0 and j > 0:
            ca[i, j] = max(c(0, j-1), distance.euclidean(P[0], Q[j]))
        elif i > 0 and j > 0:
            ca[i, j] = max(
                min(c(i-1, j), c(i-1, j-1), c(i, j-1)), distance.euclidean(P[i], Q[j])
            )
        return ca[i, j]
    
    return c(len(P) - 1, len(Q) - 1)

dtw_scores = []
frechet_scores = []

for i, row in enumerate(asoh.itertuples(index=False)):
    x_values = row[:13]
    y_values = row[13:26]
    
    x_values_normalized = [(x - x_min) / (x_max - x_min) for x in x_values]
    y_values_normalized = [(y - y_min) / (y_max - y_min) for y in y_values]
    y_values_normalized = [-1 * i for i in y_values_normalized]
    
    curve = np.column_stack((x_values_normalized, y_values_normalized))
    
    if i == 0:
        reference_curve = curve  
    
    # Compute DTW distance
    dtw_dist, _ = fastdtw(curve, reference_curve, dist=distance.euclidean)
    dtw_scores.append(dtw_dist)
    
    # Compute Fréchet distance
    frechet_dist = frechet_distance(curve, reference_curve)
    frechet_scores.append(frechet_dist)
    plt.plot(x_values_normalized, y_values_normalized, marker='x')
plt.title("Normalized Curves")
plt.xlabel("X Normalized")
plt.ylabel("Y Normalized")
plt.show()
print(f"Mean DTW Distance: {np.mean(dtw_scores)}")
print(f"Mean Fréchet Distance: {np.mean(frechet_scores)}")
```
